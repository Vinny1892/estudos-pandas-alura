### Cria uma series boleana com base  no kernel passado pela mesma (variavel residencia)

```python
residencial = ['Quitinete', 'Casa','Apartamento','Casa de Condomínio']
kernel_residencial = data['Tipo'].isin(residencial)
kernel_residencial
```

### Seleciona os itens com base na series gerada pelo kernel e reconstroi o  indice do dataframe

```python
data_residencial = data[kernel_residencial]
data_residencial.index = range(1, data_residencial.shape[0]+1)
data_residencial
```

### Retirando itens em que a coluna "Condominio" é nulo e  "Tipo" é igual a Apartamento


```python

selecao = (data_residencial["Tipo"] == "Apartamento") & (data_residencial["Condominio"].isnull())
A = data_residencial.shape[0]
data_residencial = data_residencial[~selecao]
B =  data_residencial.shape[0]
A - B
```

```python
data_residencial_filter = data_residencial.fillna({'Condominio':0 , 'IPTU': 0})
data_residencial_filter.info()
```
