```
import pandas as pd
import numpy as np
```

# Importação da base dedados e  ler os 10 primeiros registros 

```
data = pd.read_csv('./dataset/aluguel.csv' ,sep=";")
data.head(10)
```

# Colocando os tipos em um dataframe

<p> a coluna variavel é o indice do dataframe</p>

```
 data_type =  pd.DataFrame(data.dtypes ,columns=['Tipo de Dados' ])
 data_type.columns.name = "Variaveis" 
 data_type
```

# Selecionando a primeira coluna

```
tipo_imovel = data['Tipo']
tipo_imovel
```

# Retirando dados duplicados
<p> <code>Inplace</code> altera a propria instancia do objeto fazendo com que não seja presciso atribuir a uma nova variavel, removendo a imutabilidade do pandas </p>

```
tipo_imovel.drop_duplicates(inplace=True)
tipo_imovel
```

#Obtendo o tamanho da Series e Dataframe

```
print('Quantidade de linhas da Series ', tipo_imovel.shape[0])
```

```
print('Quantidade de linhas:' , data.shape[0] , ' e colunas:', data.shape[1] , 'no Dataframe')
```

## Definindo um index novo usando a função range que gera numeros do segundo parametro 'start' até o 'stop' ultimo parametro, o primeiro 'step' , é de quanto em quanto será interavel o range

```
tipo_imovel.index = range(1,tipo_imovel.shape[0]+1)
tipo_imovel
```

# Gerando um data frame a partir de uma series existente

```
df_tipo_imovel = pd.DataFrame(tipo_imovel)
df_tipo_imovel
```

# Renomeando a coluna de varias formas

```
#df_tipo_imovel.columns = ['Tipo Imovel']
#df_tipo_imovel.rename(columns={'Tipo':'Tipo Imovel'} , inplace=True)
df_tipo_imovel.rename({'Tipo':'Tipo Imovel'} , inplace=True,axis=1)
df_tipo_imovel
```
